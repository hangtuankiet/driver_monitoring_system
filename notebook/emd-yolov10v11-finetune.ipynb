```python
# YOLOv10 fine-tune
!yolo train \
    model=/kaggle/working/yolov10/yolov10n.pt \
    data=/kaggle/working/data.yaml \
    epochs=100 \
    imgsz=640 \
    batch=16 \
    project=/kaggle/working/runs/train \
    name=yolov10_custom \
    device=0
```
